```python
import requests
from bs4 import BeautifulSoup

# Send a GET request to the webpage
url = 'https://www.espn.com/mens-college-basketball/stats/player'
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the webpage
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the table containing player statistics
    table = soup.find('table', class_='Table Table--align-right Table--fixed Table--fixed-left')

    if table:
        # Extract data from the table
        headers = [header.text.strip() for header in table.find_all('th')]
        players_data = []
        for row in table.find_all('tr'):
            player_info = [data.text.strip() for data in row.find_all('td')]
            if player_info:  # Ensure it's not a header row
                players_data.append(player_info)

        # Print the headers and some player data for testing
        print(headers)
        print(players_data[:5])  # Print data for the first 5 players as an example
    else:
        print("Table not found on the webpage.")
else:
    print("Failed to retrieve the webpage.")
```

```python
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as pd

# Initialize WebDriver
driver = webdriver.Chrome()  # Assuming you are using Chrome
base_url = 'https://www.espn.com/mens-college-basketball/stats/player'
url = base_url

# Assume driver is already initialized
driver.get(url)

# Define a function to click the "Show More" button using JavaScript
def click_show_more():
    js_script = "document.querySelector('.loadMore__link').click();"
    driver.execute_script(js_script)
    time.sleep(5)  # Wait for data to load

# Click the "Show More" button multiple times to load all data
num_clicks = 14  # Number of times to click "Show More"
for _ in range(num_clicks):
    click_show_more()

# Get the page source after loading all data
page_source = driver.page_source

# Close the Selenium WebDriver
driver.quit()

# Parse the page source with BeautifulSoup
soup = BeautifulSoup(page_source, 'html.parser')

# Find the table containing player stats
table = soup.find('table', class_='Table Table--align-right')

# Initialize lists to store data
rows_data = []

# Define the columns based on the header row
columns = [header.text for header in table.find_all('th', class_='Table__TH')]

# Find all rows in the table body
rows = table.find('tbody').find_all('tr', class_='Table__TR')

# Loop through each row and extract data
for row in rows:
    row_data = [cell.text.strip() for cell in row.find_all('td', class_='Table__TD')]
    rows_data.append(row_data)

# Create a DataFrame
df = pd.DataFrame(rows_data, columns=columns)

# Display the DataFrame
print(len(df))

df.head()
```
